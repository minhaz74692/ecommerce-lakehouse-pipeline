#Bronze.Products Silver Transformation

```python
from pyspark.sql.functions import row_number, col, current_timestamp
from pyspark.sql.window import Window
```

```python
df = spark.table("bronze.products").select("*")
```

```python
df.count()
```

```python
df.limit(5).display()
```

```python
w = Window.partitionBy("product_id").orderBy(col("_load_timestamp").desc())
df = df.withColumn("rn", row_number().over(w)).filter(col("rn") == 1).drop("rn")
```

```python
df = df.filter(col("product_id").isNotNull())

df.count()
```

```python
df.write.format("delta").mode("overwrite").option("inferSchema", True).saveAsTable("silver.products")
```
