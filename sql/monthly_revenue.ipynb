```sql
%sql
-- Monthly revenue trend
-- Business question: how does product revenue move month over month?
-- Revenue = SUM(price). Freight excluded (shipping pass-through, not merchant margin).
-- Returns: one row per calendar month, chronological.
SELECT
    dd.year_month AS month,
    SUM(fs.price) AS product_revenue
FROM gold.fact_sales AS fs
JOIN gold.dim_date  AS dd
    ON fs.date_key = dd.date_key
WHERE fs.order_status = "delivered"
GROUP BY dd.year_month
ORDER BY dd.year_month;
```
