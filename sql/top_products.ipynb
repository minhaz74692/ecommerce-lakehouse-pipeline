```sql
%sql
-- Top products by revenue
-- Business question: which products drive the most product revenue?
-- Revenue = SUM(price), delivered orders only. Also returns units_sold.
-- Grain: per product (product_sk). Returns top 10.
WITH sales_products AS (
    SELECT  
        fs.product_sk, 
        COUNT(*) as units_sold, 
        SUM(fs.price) AS product_revenue
    FROM ecommerce_lakehouse.gold.fact_sales fs
    WHERE fs.order_status = 'delivered'
    GROUP BY fs.product_sk
)
SELECT
    sp.product_sk,
    dp.product_id,
    dp.product_category_name,   
    sp.units_sold,             
    sp.product_revenue
FROM sales_products sp
INNER JOIN ecommerce_lakehouse.gold.dim_product dp
    ON sp.product_sk = dp.product_sk
ORDER BY sp.product_revenue DESC
LIMIT 10;
```
